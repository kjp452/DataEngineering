```python
"""
for Loops
"""

nums = [1, 2, 3, 4, 5]

for num in nums:
    print(num)

print("")

#range

for i in range(10):
    print(i)
```

```python
"""
continue Keyword
"""

nums = [1, 2, -1, -2, 3, 4, -5, -6]

for num in nums:
    if num < 0:
        continue
    print(num)
```

```python
"""
break Keyword
"""

game_plays = ['stage1', 'stage2', 'stage3', 'gameover', 'stage4', 'stage5']

for play in game_plays:
    
    if play == 'gameover':
        print("Game is Over...")
        break
        
    print(play + " is complete. ")
```

```python
"""
List Comprehension
"""

one_to_ten = [i+1 for i in range(10)]
print(one_to_ten)
```

```python
"""
List comprehension including conditions
"""

hotspurs = [['손흥민', 'A+'], ['해리 케인', 'F'],
            ['탕기 은돔벨레', 'F'], ['델리 알리', 'B'],
            ['루카스 모우라', 'B+'], ['이메르송', 'D']]

poors = [player[0] for player in hotspurs if player[1] == 'A+']
print(poors)
```

```python
"""
Nested Loops
"""
#구구단 프로그램(1~9)
one_to_nine = [i+1 for i in range(9)]

for i in one_to_nine:
    for j in one_to_nine:
        print("%d X %d = %d" % (i, j, i*j))
```

```python
"""
Infinite Loops
"""

x = 1
y = 1
while x < 5:
    print("Do Something")
    y += 1
```
